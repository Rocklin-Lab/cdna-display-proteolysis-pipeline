```python
# import libaries

import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
#import import_ipynb
#import DataS18_STEP1_module
import STEP1_module
```

```python
## get overlapped sequences
df_overlap12 = pd.read_csv('./Raw_NGS_counts_overlapped_seqs_STEP1_lib1_lib2.csv')
df_overlap23 = pd.read_csv('./Raw_NGS_counts_overlapped_seqs_STEP1_lib2_lib3.csv')
df_overlap14 = pd.read_csv('./Raw_NGS_counts_overlapped_seqs_STEP1_lib1_lib4.csv')
df_overlap24 = pd.read_csv('./Raw_NGS_counts_overlapped_seqs_STEP1_lib1_lib3.csv')
df_overlap34 = pd.read_csv('./Raw_NGS_counts_overlapped_seqs_STEP1_lib1_lib4.csv')
```

```python
# make array including trypsin data and calculate total counts for each condition
# countsX: observed NGS counts [# of sequences, # of conditions (48)]
# # of conditions = 12 protease concentrations x 2 replicates x 2 libraries (because we are analyzing overlapped sequences between two libraries)
counts1 = np.array(np.concatenate([df_overlap12.iloc[:,3:27],df_overlap12.iloc[:,51:75]],axis=1).T)
counts2 = np.array(np.concatenate([df_overlap23.iloc[:,3:27],df_overlap23.iloc[:,51:75]],axis=1).T)
counts3 = np.array(np.concatenate([df_overlap14.iloc[:,3:27],df_overlap14.iloc[:,51:75]],axis=1).T)
counts4 = np.array(np.concatenate([df_overlap24.iloc[:,3:27],df_overlap24.iloc[:,51:75]],axis=1).T)
counts5 = np.array(np.concatenate([df_overlap34.iloc[:,3:27],df_overlap34.iloc[:,51:75]],axis=1).T)

# run the module
samples = STEP1_module.protease_calibration(counts1,counts2,counts3,counts4,counts5)

# save protease concentrations
eff_pro_con = {'protease_v1':np.median(samples['protease_v1'],axis=0),
               'protease_v2':np.median(samples['protease_v2'],axis=0),
               'protease_v3':np.median(samples['protease_v3'],axis=0),
               'protease_v4':np.median(samples['protease_v4'],axis=0)}
import pickle
with open('Effective_protease_tryp_related_DataS20_221104', 'wb') as p:
    pickle.dump(eff_pro_con, p)
    
df_tryspin_con = pd.DataFrame()
df_tryspin_con['protease_v1'] = np.median(samples['protease_v1'],axis=0)
df_tryspin_con['protease_v2'] = np.median(samples['protease_v2'],axis=0)
df_tryspin_con['protease_v3'] = np.median(samples['protease_v3'],axis=0)
df_tryspin_con['protease_v4'] = np.median(samples['protease_v4'],axis=0)
df_tryspin_con.to_csv('Effective_protease_tryp.csv')
df_tryspin_con
```
