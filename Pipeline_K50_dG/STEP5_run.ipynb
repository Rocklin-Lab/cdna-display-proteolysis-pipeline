```python
# import libaries

import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import import_ipynb
import STEP5_module
```

```python
###############################
#
#  Calculation of K50 and dG for Lib1
#
###############################

# get NGS counts
df = pd.read_csv('../Raw_NGS_count_tables/NGS_count_lib1.csv')

# unfolded Ks are predicted in STEP4 
df_unfold = pd.read_csv('./Lib1_unfolded.csv')
df = pd.merge(df,df_unfold,on=['name','dna_seq'],how='left')
df = df[(df['v1_T01']>0)&(df['v1_T13']>0)&(df['v1_C01']>0)&(df['v1_C13']>0)&(df['v1_T01']<5000)].reset_index(drop=True) 

# run numpyro model
samples1 = STEP5_module.obtain_samples('v1',df)

# summarize the result from numpyromodel
dfsum = STEP5_module.samples2sumdf(samples1,'v1',df)
dfsum.to_csv('Lib1_K50dG.csv',index=False)
dfsum
```
