```python
###############################
#
#  Set up a model of amino acid classification model
#
###############################

def aa_classification_model(dGs, categories):
    # dGs: folding stability [# of amino acids (19; without Cystein), # of sites]
    # categories: observed wildtype amino acids (one hot vectors) [# of amino acids (19), # of sites]
    
    
    # response_centers: 100 thresholds for 100 logistic functions for activation (kcal/mol) [100]
    # response_amplitudes: amplitude for 100 logistic functions [100]
    # The parameters are shared between 19 amino acids
    response_centers = np.linspace(-2,7,100) 
    response_amplitudes = jax.numpy.exp(numpyro.sample("response_amplitudes", dist.Normal(np.resize(0, np.shape(response_centers)), 1  ))) *0.1
    
    # response: relative wild-type possibilities before adding amino-acid-specific offsets calculated from dG values for each site [# of amino acids (19), # of sites]
    logistic_steepness = 5
    response = jax.numpy.sum(jax.scipy.special.expit((jax.numpy.expand_dims(dGs, -1) -response_centers)*logistic_steepness  ) * response_amplitudes, axis=-1) 
    
    # logit_offset: offset for each specific amino acid [# of amino acids -1(18)]
    # Sampled 18 offsets because they are relative values (i.e. we fixed the offest for the Pro (the last amino acid))
    logit_offset = numpyro.sample("logit_offset", dist.Normal(np.resize(0, (1,18)), 2))

    # probs: relative wild-type possibilities after adding amino-acid-specific offsets  [# of amino acids (19), # of sites]
    # probs = Softmax(response + offset)
    probs=jax.nn.softmax(response+jax.numpy.pad(logit_offset, ((0,0),(0,1))),axis=-1)
    
    # fitting parameters assuming that amino acids with max probs match observed wild-type amino acids
    numpyro.sample("obs_aa", dist.Categorical(probs=probs), obs = categories)
```

```python
###############################
#
#  Preparation of input data
#
###############################

# import the data
df = pd.read_csv('../Data_tables_for_figs/dG_non_redundant_natural_Fig5.csv')
df
```

```python
# Split the dataset into train and test dataset
domain_list = list(set(df['pdb_name']))
domain_list.sort()

train_domain_list = domain_list[:85]

df_train, df_test = pd.DataFrame(),pd.DataFrame()
for i in range(len(df)):
    if df['pdb_name'][i] in train_domain_list:
        df_train = df_train.append(df.iloc[i,:])
    else:
        df_test = df_test.append(df.iloc[i,:])
        
df_train = df_train.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
df_train
```

```python
# aas: list of 19 amino acids
aas=list(df_train.columns[3:23])

# dGs: folding stability [# of amino acids (19), # of sites]
dGs = df_train[df_train.columns[3:23]].to_numpy()

# cats_array: observed wildtype amino acids (one hot vectors) [# of amino acids (19), # of sites]
cats=np.array([aas.index(x) for x in df_train['wt_aa']])
cats_array=np.zeros(np.shape(dGs))
for i, idx in enumerate(cats):
    cats_array[i, idx] = 1
cats_array
```

```python
# run the model
rng_key = random.PRNGKey(0)
rng_key, rng_key_ = random.split(rng_key)

kernel = NUTS(aa_classification_model)
mcmc = MCMC(kernel, num_warmup=500, num_samples=100)
mcmc.run(rng_key_, dGs = dGs, categories = cats)
samples_1 = mcmc.get_samples()
```

```python
###############################
#
#  Output check
#
###############################

df_offset = pd.DataFrame()
df_offset['aa'] = list(aas)
df_offset['offset'] = ((list(np.median(samples_1['logit_offset'],axis=0)[0])+[0])-np.mean((list(np.median(samples_1['logit_offset'],axis=0)[0])+[0])))/np.log(2)
df_offset = df_offset.sort_values('offset',ascending=False).reset_index(drop=True)
df_offset
```

```python
# Calculate relative probability of wild-type amino acids
response_centers = np.linspace(-2,7,100) 
response_amplitudes = jax.numpy.exp(np.median(samples_1['response_amplitudes'],axis=0)) *0.1
logit_offset=np.median(samples_1['logit_offset'],axis=0)
dGs = df_train[df_train.columns[3:23]].to_numpy() ##
response = jax.numpy.sum(jax.scipy.special.expit((jax.numpy.expand_dims(dGs, -1) -response_centers)*5  ) * response_amplitudes, axis=-1) - jax.numpy.sum(jax.scipy.special.expit((1-response_centers)*5) * response_amplitudes) 
train_probs = jax.nn.softmax(response+jax.numpy.pad(logit_offset, ((0,0),(0,1))),axis=-1)

# Obsrved wild-type amino acids
true_label = np.eye(19)[[int(aas.index(x)) for x in df_train['wt_aa']]] 

# Calculate PPL and recovery rate
df_response = pd.DataFrame(train_probs)
df_response['pred_index'] = [list(x).index(max(x)) for x in train_probs]
df_response['true_index'] = [list(x).index(max(x)) for x in true_label]

PPL, Recovery_rate = np.exp(-sum(sum(np.array(true_label*np.log(df_response.iloc[:,:19]))))/len(df_response)),len(df_response[df_response['pred_index']==df_response['true_index']])/len(df_response)

print('PPL:'+str(PPL)+', recovery rate:'+str(Recovery_rate))
```
