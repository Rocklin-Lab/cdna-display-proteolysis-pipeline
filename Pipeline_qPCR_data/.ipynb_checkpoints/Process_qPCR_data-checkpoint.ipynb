```python
df_qpcr = pd.read_csv('./Raw_qPCR_data_FigS1.csv')
df_qpcr
```

```python
# get raw qPCR data

df_qpcr_t = df_qpcr[df_qpcr['protease']=='trypsin']
df_qpcr_c = df_qpcr[df_qpcr['protease']=='chymotrypsin']
df_qpcr_c
```

```python
# get protease concentration
protease_con = np.array(df_qpcr_c['protease_con'])
protease_con
```

```python
###############################
#
#  Set up a model to sample kmax*t, and K50 
#
###############################

def qpcr2k_model(obs_survival):
    # n: number of sequences
    n=8
    
    # log10_K50: log10 K50 values for each sequence [# of sequences], sampled in wide normal distribution
    log10_K50 = numpyro.sample("log10_K50", dist.Normal(np.resize(0,n), 4) ) 
    
    # log10_kamx_times_t: log10 kmax*t  for each sequence [# of sequences], sampled in wide normal distribution
    log10_kmax_times_t = numpyro.sample("log10_kmax_times_t", dist.Normal(np.resize(0,n), 4) ) 

    # survival: relative ratio of each sequence for each condition to initial condition (no protease) [# of sequences (n), # of conditions (12)]
    # survival = exp(- kmax*t*[protease]/(K50+[protease]))
    survival = jax.numpy.log10(jax.numpy.exp(-jax.numpy.outer((10**log10_kmax_times_t), protease_con)/((jax.numpy.resize(10.0**log10_K50,(12,n)).T)+jax.numpy.resize(protease_con,(n,12)))))
    
    ## fitting parameters using the observed fraction of cDNA
    # sigma: uncertainty of fractions of cDNA
    sigma = numpyro.sample("sigma", dist.Exponential(1)) 
    obs_survival = numpyro.sample("obs_survival",dist.TransformedDistribution(dist.Normal(0, 1), dist.transforms.AffineTransform(survival,sigma)), obs = obs_survival)
```

```python
###############################
#
#  Run the model for trypsin data & draw a plot
#
###############################

rng_key = random.PRNGKey(1)
rng_key, rng_key_ = random.split(rng_key)

kernel = NUTS(qpcr2k_model, init_strategy=initialization.init_to_feasible())
mcmc = MCMC(kernel, num_warmup=100, num_samples=50, num_chains=1)

mcmc.run(rng_key_, obs_survival=np.array(np.log10(2**df_qpcr_t.iloc[:,1:9])).T)

samples_tryp_2para=mcmc.get_samples()

for i in range(8):
    plt.scatter(samples_tryp_2para['log10_K50'][:,i],samples_tryp_2para['log10_kmax_times_t'][:,i],alpha=0.25)
plt.xlabel('log10_K50_trypsin')
plt.ylabel('log10_kmax_times_t')
```

```python
###############################
#
#  Run the model for chymotrypsin data & draw a plot
#
###############################

rng_key = random.PRNGKey(1)
rng_key, rng_key_ = random.split(rng_key)

kernel = NUTS(qpcr2k_model, init_strategy=initialization.init_to_feasible())
mcmc = MCMC(kernel, num_warmup=100, num_samples=50, num_chains=1)

mcmc.run(rng_key_, obs_survival=np.array(np.log10(2**df_qpcr_c.iloc[:,1:9])).T)
samples_chymo_2para=mcmc.get_samples()

for i in range(8):
    plt.scatter(samples_chymo_2para['log10_K50'][:,i],samples_chymo_2para['log10_kmax_times_t'][:,i],alpha=0.25)
plt.xlabel('log10_K50_chymotrypsin')
plt.ylabel('log10_kmax_times_t')
```

```python
###############################
#
#  Set up a model to sample and K50 with fixed kmax*t
#
###############################

def qpcr2k_model_kmaxfixed(obs_survival):
    # n: number of sequences
    n=8
    
    # log10_K50: log10 K50 values for each sequence [# of sequences], sampled in wide normal distribution
    log10_K50 = numpyro.sample("log10_K50", dist.Normal(np.resize(0,n), 4) ) 
    
    # log10_kamx_times_t: fixed values based on the above analysis
    log10_kmax_times_t = 0.65

    # survival: relative ratio of each sequence for each condition to initial condition (no protease) [# of sequences (n), # of conditions (12)]
    # survival = exp(- kmax*t*[protease]/(K50+[protease]))
    survival = jax.numpy.log10(jax.numpy.exp(-jax.numpy.outer((10**log10_kmax_times_t), protease_con)/((jax.numpy.resize(10.0**log10_K50,(12,n)).T)+jax.numpy.resize(protease_con,(n,12)))))
    
    ## fitting parameters using the observed fraction of cDNA
    # sigma: uncertainty of fractions of cDNA
    sigma = numpyro.sample("sigma", dist.Exponential(1)) 
    obs_survival = numpyro.sample("obs_survival",dist.TransformedDistribution(dist.Normal(0, 1), dist.transforms.AffineTransform(survival,sigma)), obs = obs_survival)
```

```python
###############################
#
#  Run the model for trypsin data & draw a plot
#
###############################

rng_key = random.PRNGKey(1)
rng_key, rng_key_ = random.split(rng_key)

kernel = NUTS(qpcr2k_model_kmaxfixed, init_strategy=initialization.init_to_feasible())
mcmc = MCMC(kernel, num_warmup=100, num_samples=50, num_chains=1)

mcmc.run(rng_key_, obs_survival=np.array(np.log10(2**df_qpcr_t.iloc[:,1:9])).T)
samples_tryp_1para=mcmc.get_samples()

K50_T_list = []
for i in range(8):
    sns.kdeplot(samples_tryp_1para['log10_K50'][:,i],alpha=1)
    K50_T_list.append(np.median(samples_tryp_1para['log10_K50'][:,i]))
plt.xlabel('log10_K50_trypsin')
K50_T_list
```

```python
###############################
#
#  Run the model for chymotrypsin data & draw a plot
#
###############################

rng_key = random.PRNGKey(1)
rng_key, rng_key_ = random.split(rng_key)

kernel = NUTS(qpcr2k_model_kmaxfixed, init_strategy=initialization.init_to_feasible())
mcmc = MCMC(kernel, num_warmup=100, num_samples=50, num_chains=1)

mcmc.run(rng_key_, obs_survival=np.array(np.log10(2**df_qpcr_c.iloc[:,1:9])).T)
samples_chymo_1para=mcmc.get_samples()


K50_C_list = []
for i in range(8):
    sns.kdeplot(samples_chymo_1para['log10_K50'][:,i],alpha=0.7)
    K50_C_list.append(np.median(samples_tryp_1para['log10_K50'][:,i]))
plt.xlabel('log10_K50_chymotrypsin')
K50_C_list
```
